```python
import sys
from dwcal import delay_weighted_cal as dwcal
import numpy as np
import pyuvdata
import matplotlib.pyplot as plt
import matplotlib
import scipy.optimize
```

```python
data, model = dwcal.get_test_data(
    model_path="/Users/ruby/Astro/FHD_outputs/fhd_rlb_model_GLEAM_bright_sources_Dec2021",
    model_use_model=True,
    data_path="/Users/ruby/Astro/FHD_outputs/fhd_rlb_model_GLEAM_bright_sources_Dec2021",
    data_use_model=False,
    obsid="1061316296",
    pol="XX",
    use_autos=False,
    debug_limit_freqs=None,
    use_antenna_list=None,
    use_flagged_baselines=False,
)
```

```python
diff_vis = data.diff_vis(model, inplace=False)
del(data)
del(model)
```

```python
print(np.shape(diff_vis.data_array))
```

```python
delay_array = np.fft.fftfreq(diff_vis.Nfreqs, d=diff_vis.channel_width)
delay_array = np.fft.fftshift(delay_array)
diff_fft_abs = np.abs(np.fft.fftshift(np.fft.fft(diff_vis.data_array, axis=2), axes=2))
```

```python
np.shape(diff_fft_abs)
```

```python
bl_lengths = np.sqrt(np.sum(diff_vis.uvw_array ** 2.0, axis=1))
nbins=100
bl_bin_edges = np.linspace(np.min(bl_lengths), np.max(bl_lengths), num=nbins+1)
binned_delay_spec = np.full([nbins, diff_vis.Nfreqs], np.nan, dtype="float")
for bin_ind in range(nbins):
    bl_inds = np.where((bl_lengths > bl_bin_edges[bin_ind]) & (bl_lengths <= bl_bin_edges[bin_ind+1]))[0]
    if len(bl_inds) > 0:
        binned_delay_spec[bin_ind, :] = np.sum(diff_fft_abs[bl_inds, 0, :, 0], axis=0)/len(bl_inds)
```

```python
c = 3.0 * 10 ** 8  # Speed of light

use_cmap = matplotlib.cm.get_cmap('plasma')
use_cmap.set_bad(color='whitesmoke')
plt.imshow(
    binned_delay_spec.T, origin='lower', interpolation='none', cmap=use_cmap, 
    norm=matplotlib.colors.LogNorm(vmin=1e-1, vmax=1e3),
    extent=[np.min(bl_bin_edges), np.max(bl_bin_edges), np.min(delay_array),
                np.max(delay_array)],
    aspect='auto',
)
plt.plot(
    [np.min(bl_bin_edges), np.max(bl_bin_edges)],
    [np.min(bl_bin_edges)/c, np.max(bl_bin_edges)/c],
    "--",
    color="white",
    linewidth=1,
)
plt.plot(
    [np.min(bl_bin_edges), np.max(bl_bin_edges)],
    [np.min(bl_bin_edges)/c, -np.max(bl_bin_edges)/c],
    "--",
    color="white",
    linewidth=1,
)
plt.plot(
    [np.min(bl_bin_edges), np.max(bl_bin_edges)],
    [np.min(bl_bin_edges)/c*0.20, np.max(bl_bin_edges)/c*0.20],
    "--",
    color="white",
    linewidth=1,
)
plt.plot(
    [np.min(bl_bin_edges), np.max(bl_bin_edges)],
    [np.min(bl_bin_edges)/c*0.20, -np.max(bl_bin_edges)/c*0.20],
    "--",
    color="white",
    linewidth=1,
)
cbar = plt.colorbar()
cbar.ax.set_ylabel(
    '', rotation=270, labelpad=15
)
plt.xlabel('Baseline Length (m)')
plt.ylim([-2e-6,2e-6])
plt.ylabel('Delay (s)')
plt.show()
```

```python
def cost_func_step_fit(x, diff_fft_abs_squeezed, delay_array, bl_lengths):
    wedge_error = x[0]
    window_error = x[1]
    boundary_slope = x[2]/c
    fit = np.full_like(diff_fft_abs_squeezed, window_error)
    for delay_ind, delay_val in enumerate(delay_array):
        wedge_bls = np.where(bl_lengths*boundary_slope > np.abs(delay_val))[0]
        if len(wedge_bls) > 0:
            fit[wedge_bls, delay_ind] = wedge_error
    return np.sum((diff_fft_abs_squeezed-fit)**2.)
```

```python
def cost_func_tukey_fit(x, diff_fft_abs_squeezed, delay_array, bl_lengths):
    wedge_amp = x[0]
    window_amp = x[1]
    inner_boundary_slope = np.abs(x[2])/c
    outer_boundary_slope = inner_boundary_slope + np.abs(x[3])/c
    fit = model_tukey_error(wedge_amp, window_amp, inner_boundary_slope, outer_boundary_slope, delay_array, bl_lengths)

    return np.sum((diff_fft_abs_squeezed-fit)**2.)
```

```python
def tukey_function(xvals, xmin, xmax, func_min, func_max):
    
    arg_scaling = np.pi/np.abs(xmax-xmin)
    tukey = np.cos(arg_scaling*(xvals-xmin))/2. + .5
    tukey *= (func_max - func_min)
    tukey += func_min
    return tukey
```

```python
vals = np.arange(10,50)
tukey = tukey_function(vals, 10, 50, .5, 3)
plt.plot(vals, tukey)
```

```python
def model_tukey_error(wedge_amp, window_amp, inner_boundary_slope, outer_boundary_slope, delay_array, bl_lengths):

    fit = np.full((len(bl_lengths), len(delay_array)), window_amp)
    
    for bl_ind, bl_length in enumerate(bl_lengths):
        wedge_delays = np.where(np.abs(delay_array) <= bl_length*inner_boundary_slope)[0]
        if len(wedge_delays) > 0:
            fit[bl_ind, wedge_delays] = wedge_amp
        intermediary_delays = np.intersect1d(
            np.where(np.abs(delay_array) <= bl_length*outer_boundary_slope)[0],
            np.where(np.abs(delay_array) > bl_length*inner_boundary_slope)[0]
        )
        if len(intermediary_delays) > 0:
            tukey_values = tukey_function(
                np.abs(delay_array[intermediary_delays]),
                bl_length*inner_boundary_slope,
                bl_length*outer_boundary_slope, 
                window_amp, 
                wedge_amp
            )
            fit[bl_ind, intermediary_delays] = tukey_values
    
    return fit
```

```python
x0 = np.array([1.,1.,1.])
result = scipy.optimize.minimize(
    cost_func_step_fit,
    x0,
    args=(
        diff_fft_abs.squeeze(),
        delay_array,
        bl_lengths,
    ),
    method = 'Nelder-Mead'
)
```

```python
x0 = np.array([1e-1,1e2,.2,1.])
result = scipy.optimize.minimize(
    cost_func_tukey_fit,
    x0,
    args=(
        diff_fft_abs.squeeze(),
        delay_array,
        bl_lengths,
    ),
    method = 'Nelder-Mead'
)
```
