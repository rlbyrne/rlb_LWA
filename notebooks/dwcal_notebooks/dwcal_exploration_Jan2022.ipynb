```python
import sys
sys.path.append('/Users/ruby/Astro/rlb_LWA/dwcal')
import delay_weighted_cal as dwcal
import numpy as np
import pyuvdata
import matplotlib.pyplot as plt
import matplotlib
```

```python
data, model = dwcal.get_test_data(
    model_path="/Users/ruby/Astro/FHD_outputs/fhd_rlb_model_GLEAM_bright_sources_Dec2021",
    model_use_model=True,
    data_path="/Users/ruby/Astro/FHD_outputs/fhd_rlb_model_GLEAM_bright_sources_Dec2021",
    data_use_model=False,
    obsid="1061316296",
    pol="XX",
    use_autos=False,
    debug_limit_freqs=None,
    use_antenna_list=None,
    use_flagged_baselines=False,
)
```

```python
data.diff_vis(model, inplace=True)
```

```python
delay_array = np.fft.fftfreq(data.Nfreqs, d=data.channel_width)
delay_array = np.fft.fftshift(delay_array)
```

```python
np.shape(data.data_array)
```

```python
delay_spec_data = np.abs(np.fft.fft(data.data_array, axis=2))**2.
delay_spec_data = np.fft.fftshift(delay_spec_data, axes=2)
```

```python
plt.plot(data.freq_array[0,:], np.abs(data.data_array[3000,0,:,0]))
```

```python
plt.plot(delay_array, delay_spec_data[3000,0,:,0], 'o', markersize=1)
```

```python
bl_lengths = np.sqrt(np.sum(data.uvw_array ** 2.0, axis=1))
```

```python
nbins=100
bl_bin_edges = np.linspace(np.min(bl_lengths), np.max(bl_lengths), num=nbins+1)
binned_delay_spec = np.full([nbins, data.Nfreqs], np.nan, dtype="float")
for bin_ind in range(nbins):
    bl_inds = np.where((bl_lengths > bl_bin_edges[bin_ind]) & (bl_lengths <= bl_bin_edges[bin_ind+1]))[0]
    if len(bl_inds) > 0:
        binned_delay_spec[bin_ind, :] = np.sum(delay_spec_data[bl_inds, 0, :, 0], axis=0)/len(bl_inds)
```

```python
plt.plot(delay_array, binned_delay_spec[90,:])
```

```python
use_cmap = matplotlib.cm.get_cmap('Spectral')
use_cmap.set_bad(color='whitesmoke')
plt.imshow(
    binned_delay_spec.T, origin='lower', interpolation='none', vmin=10, vmax=1e7, cmap=use_cmap, 
    norm=matplotlib.colors.LogNorm(),
    extent=[np.min(bl_bin_edges), np.max(bl_bin_edges), np.min(delay_array),
                np.max(delay_array)],
    aspect='auto'
)
cbar = plt.colorbar()
cbar.ax.set_ylabel(
    '', rotation=270, labelpad=15
)
plt.xlabel('Baseline Length (m)')
plt.ylabel('Delay (s)')
plt.show()
```
