```python
def newtons_method_optimizer(
    x0,
    Nants,
    Nfreqs,
    Nbls,
    model_visibilities,
    gains_exp_mat_1,
    gains_exp_mat_2,
    cov_mat,
    data_visibilities,
    step_size = .5,
    covergence_condition = 0.001,
):

    n_iters = 0
    convergence_iters = 0
    while convergence_iters < 3:
        hess_mat = dwcal.hess_dw_cal(
            x0,
            Nants,
            Nfreqs,
            Nbls,
            model_visibilities,
            gains_exp_mat_1,
            gains_exp_mat_2,
            cov_mat,
            data_visibilities,
        )
        hess_mat_inv = np.linalg.inv(hess_mat)
        del hess_mat
        jac = dwcal.jac_dw_cal(
            x0,
            Nants,
            Nfreqs,
            Nbls,
            model_visibilities,
            gains_exp_mat_1,
            gains_exp_mat_2,
            cov_mat,
            data_visibilities,
        )
        x1 = x0 - step_size*np.matmul(hess_mat_inv, jac)
        del hess_mat_inv
        del jac
        cost = dwcal.cost_function_dw_cal(
            x0,
            Nants,
            Nfreqs,
            Nbls,
            model_visibilities,
            gains_exp_mat_1,
            gains_exp_mat_2,
            cov_mat,
            data_visibilities,
        )
        print(f"Iteration {n_iters}, cost func value: {cost}")
        check_conv = np.max(np.abs(x1-x0))
        if check_conv < covergence_condition:
            convergence_iters += 1
        else:
            convergence_iters = 0
        x0 = x1
        n_iters += 1
```

```python
def lagrange_cost(
    x,
    Nants,
    Nfreqs,
    Nbls,
    model_visibilities,
    gains_exp_mat_1,
    gains_exp_mat_2,
    weight_mat,
    data_visibilities,
    verbose=True,
    lambda_val=1.0,
):

    gains = np.reshape(x, (2, Nants, Nfreqs))
    gains = gains[0, :, :] + 1.0j * gains[1, :, :]
    cost = lambda_val * np.sum(np.sum(np.angle(gains), axis=0) ** 2.0)
    return cost

def lagrange_jac(
    x,
    Nants,
    Nfreqs,
    Nbls,
    model_visibilities,
    gains_exp_mat_1,
    gains_exp_mat_2,
    weight_mat,
    data_visibilities,
    verbose=True,
    lambda_val=1.0,
):

    gains = np.reshape(x, (2, Nants, Nfreqs))
    gains = gains[0, :, :] + 1.0j * gains[1, :, :]
    grad = (
        2
        * lambda_val
        * np.conj(gains)
        / np.abs(gains) ** 2.0
        * np.sum(np.angle(gains), axis=0)[np.newaxis, :]
    )

    grad = np.stack((np.imag(grad), np.real(grad)), axis=0).flatten()
    return grad

def lagrange_hess(
    x,
    Nants,
    Nfreqs,
    Nbls,
    model_visibilities,
    gains_exp_mat_1,
    gains_exp_mat_2,
    weight_mat,
    data_visibilities,
    verbose=True,
    lambda_val=1.0,
):

    gains = np.reshape(x, (2, Nants, Nfreqs))
    gains = gains[0, :, :] + 1.0j * gains[1, :, :]
    hess = np.zeros((Nants, Nants, Nfreqs, Nfreqs, 3), dtype=float)

    im_part = np.imag(gains)/np.abs(gains)**2.
    real_part = np.real(gains)/np.abs(gains)**2.
    arg_sum = np.sum(np.angle(gains), axis=0)
    
    # Real-real derivative
    term1_rr = 2 * lambda_val * np.einsum("ik,jk->ijk", im_part, im_part)
    term2_rr = 4 * lambda_val * arg_sum * im_part * real_part
    hess[:, :, np.arange(Nfreqs), np.arange(Nfreqs), 0] = term1_rr
    for ant_ind in range(Nants):
        hess[ant_ind, ant_ind, np.arange(Nfreqs), np.arange(Nfreqs), 0] += term2_rr[ant_ind, :]
        
    # Real-imaginary derivative
    term1_ri = -2 * lambda_val *np.einsum("ik,jk->ijk", im_part, real_part)
    term2_ri = 2 * lambda_val * arg_sum * (im_part**2. - real_part**2.)
    hess[:, :, np.arange(Nfreqs), np.arange(Nfreqs), 1] = term1_ri
    for ant_ind in range(Nants):
        hess[ant_ind, ant_ind, np.arange(Nfreqs), np.arange(Nfreqs), 1] += term2_ri[ant_ind, :]
        
    # Imaginary-imaginary derivative
    term1_ii = 2 * lambda_val * np.einsum("ik,jk->ijk", real_part, real_part)
    term2_ii = -1 * term2_rr
    hess[:, :, np.arange(Nfreqs), np.arange(Nfreqs), 2] = term1_ii
    for ant_ind in range(Nants):
        hess[ant_ind, ant_ind, np.arange(Nfreqs), np.arange(Nfreqs), 2] += term2_ii[ant_ind, :]
        
    hess_reformatted = np.zeros((2, Nants * Nfreqs, 2, Nants * Nfreqs), dtype=float)
    hess_reformatted[0, :, 0, :] = np.transpose(
        hess[:, :, :, :, 0], (0, 2, 1, 3)
    ).reshape(Nants * Nfreqs, Nants * Nfreqs)
    hess_reformatted[0, :, 1, :] = np.transpose(
        hess[:, :, :, :, 1], (0, 2, 1, 3)
    ).reshape(Nants * Nfreqs, Nants * Nfreqs)
    hess_reformatted[1, :, 0, :] = np.transpose(
        hess[:, :, :, :, 1], (1, 3, 0, 2)
    ).reshape(Nants * Nfreqs, Nants * Nfreqs)
    hess_reformatted[1, :, 1, :] = np.transpose(
        hess[:, :, :, :, 2], (0, 2, 1, 3)
    ).reshape(Nants * Nfreqs, Nants * Nfreqs)
    del hess
    hess_reformatted = hess_reformatted.reshape(2 * Nants * Nfreqs, 2 * Nants * Nfreqs)
    
    return hess_reformatted
```
