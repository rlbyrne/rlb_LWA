```python
def newtons_method_optimizer(
    x0,
    Nants,
    Nfreqs,
    Nbls,
    model_visibilities,
    gains_exp_mat_1,
    gains_exp_mat_2,
    cov_mat,
    data_visibilities,
    step_size = 1.,
    covergence_condition = 0.001,
):

    n_iters = 0
    convergence_iters = 0
    while convergence_iters < 3:
        hess_mat = dwcal.hess_dw_cal(
            x0,
            Nants,
            Nfreqs,
            Nbls,
            model_visibilities,
            gains_exp_mat_1,
            gains_exp_mat_2,
            cov_mat,
            data_visibilities,
        )
        hess_mat_inv = np.linalg.inv(hess_mat)
        del hess_mat
        jac = dwcal.jac_dw_cal(
            x0,
            Nants,
            Nfreqs,
            Nbls,
            model_visibilities,
            gains_exp_mat_1,
            gains_exp_mat_2,
            cov_mat,
            data_visibilities,
        )
        x1 = x0 - step_size*np.matmul(hess_mat_inv, jac)
        del hess_mat_inv
        del jac
        cost = dwcal.cost_function_dw_cal(
            x0,
            Nants,
            Nfreqs,
            Nbls,
            model_visibilities,
            gains_exp_mat_1,
            gains_exp_mat_2,
            cov_mat,
            data_visibilities,
        )
        print(f"Iteration {n_iters}, cost func value: {cost}")
        check_conv = np.max(np.abs(x1-x0))
        if check_conv < covergence_condition:
            convergence_iters += 1
        else:
            convergence_iters = 0
        x0 = x1
        n_iters += 1
```

```python
def grad_descent_optimizer(
    x0,
    Nants,
    Nfreqs,
    Nbls,
    model_visibilities,
    gains_exp_mat_1,
    gains_exp_mat_2,
    cov_mat,
    data_visibilities,
    init_step_size = .0001,
    covergence_condition = 0.001,
):

    n_iters = 0
    convergence_iters = 0
    while convergence_iters < 3:
        jac = dwcal.jac_dw_cal(
            x0,
            Nants,
            Nfreqs,
            Nbls,
            model_visibilities,
            gains_exp_mat_1,
            gains_exp_mat_2,
            cov_mat,
            data_visibilities,
        )
        print(jac)
        if n_iters == 0:
            step_size = init_step_size
        else:
            step_size = np.abs(np.matmul((x1-x0).T, jac-jac_prev))/np.sum((jac-jac_prev)**2.)
        x1 = x0 - step_size*jac
        cost = dwcal.cost_function_dw_cal(
            x0,
            Nants,
            Nfreqs,
            Nbls,
            model_visibilities,
            gains_exp_mat_1,
            gains_exp_mat_2,
            cov_mat,
            data_visibilities,
        )
        print(f"Iteration {n_iters}, cost func value: {cost}")
        check_conv = np.max(np.abs(x1-x0))
        if check_conv < covergence_condition:
            convergence_iters += 1
        else:
            convergence_iters = 0
            
        jac_prev = jac
        x0 = x1
        n_iters += 1
```
